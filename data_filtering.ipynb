剔除无摘要数据后以及过滤type和language的references和citations：

```python
references = references[(references.abstract_inverted_index.notna())&(references.abstract_inverted_index!='{}')].reset_index(drop=True)
citations = citations[(citations.abstract_inverted_index.notna())&(citations.abstract_inverted_index!='{}')].reset_index(drop=True)
```

```python
references = references[references.type.isin(['article','book-chapter','review'])].reset_index(drop=True)
citations = citations[citations.type.isin(['article','book-chapter','review'])].reset_index(drop=True)
```

```python
references = references[references.language == 'en'].reset_index(drop=True)
citations = citations[citations.language == 'en'].reset_index(drop=True)
```

```python
print(len(references), len(citations))
```

```python
import ast
from tqdm import trange
abstracts = []
for i in trange(len(references)):
    word_indices = ast.literal_eval(references.abstract_inverted_index[i])
    # 第一步：创建一个反向映射字典，将索引映射到单词
    index_to_word = {}
    for word, indices in word_indices.items():
        for index in indices:
            index_to_word[index] = word

    # 第二步：确定摘要的总长度（最大索引+1）
    max_index = max(index_to_word.keys())
    total_length = max_index + 1

    # 第三步：按顺序构建摘要
    abstract_words = []
    for i in range(total_length):
        if i in index_to_word:
            abstract_words.append(index_to_word[i])
        else:
            abstract_words.append('[MISSING]')  # 处理可能的缺失索引

    # 第四步：将单词列表拼接成完整摘要
    abstract = ' '.join(abstract_words)
    abstracts.append(abstract)
references['abstract'] = abstracts
```
