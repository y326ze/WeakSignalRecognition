```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import Rectangle
```

```python
# 2020
years = range(2020, 2025)
words = ['blockchain','transaction','application','teaching','environment']
weakness = [[0.0869,0.9414,1,0.7478,0.8072],[0.0869,1,1,0.7478,0.4072],[0.0869,0.9414,1,1,1],[0.0869,0.1015,0.2197,0.2704,0.1915],[0.0869,0.1701,0.2566,0.4043,0.4072]]
docs = [[25,135,100,86,134],[17,48,29,44,32],[88,149,215,233,283],[10,8,34,7,12],[31,47,68,90,130]]

# 创建画布和子图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# --------------------- 左子图：折线图 ---------------------
# 颜色和形状配置
line_config = {
    words[0]:   {'color': '#ae017e', 'marker': 'o', 'size': 10},  # 圆形
    words[1]:   {'color': '#6a51a3', 'marker': 's', 'size': 9},  # 方形
    words[2]:   {'color': '#d94701', 'marker': '^', 'size': 13},  # 上三角
    words[3]:   {'color': '#41ab5d', 'marker': 'h', 'size': 12},  # 六边形
    words[4]:   {'color': '#2171b5', 'marker': '*', 'size': 18}   # 星形
}

for word in words:
    idx = words.index(word)
    ax1.plot(years, weakness[idx], 
             color=line_config[word]['color'],
             marker=line_config[word]['marker'],
             label=word,
             alpha=0.5,
             linestyle='--',
             linewidth=2.5,
             markersize=line_config[word]['size'],
             markeredgecolor='white',
             markerfacecolor=line_config[word]['color'],
             markeredgewidth=1.5)

# 在y=0.1和y=0.01处添加横向虚线
ax1.axhline(y=0.1, color='gray', linestyle='--', linewidth=2, alpha=0.7)
ax1.axhline(y=0.01, color='gray', linestyle='--', linewidth=2, alpha=0.7)

# 添加高度为0.1的长方形（示例：覆盖2020-2022年）
rect = Rectangle((years[0]-0.2, 0.01), width=years[-1]-years[0]+0.4, height=0.09, 
                 facecolor='red', alpha=0.1)
ax1.add_patch(rect)

ax1.set_title('Weakness Trend', fontsize=15)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Weakness Value', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(loc='upper left', fontsize=9, framealpha=0.9)
ax1.set_xticks(years)
ax1.set_xlim(years[0]-0.2,years[-1]+0.2)
ax1.set_ylim(0,None)


# --------------------- 右子图：分块热力图（修正行高） ---------------------
docs_matrix = np.array(docs)

# 为每行定义不同的颜色映射（顺序已反转）
row_cmaps = [
    LinearSegmentedColormap.from_list(words[4], ['#fff7fb', '#e7298a']),    # 粉
    LinearSegmentedColormap.from_list(words[3], ['#fcfbfd', '#9e9ac8']),    # 紫
    LinearSegmentedColormap.from_list(words[2], ['#fff5eb', '#fd8d3c']),    # 橙
    LinearSegmentedColormap.from_list(words[1], ['#f7fcf5', '#74c476']),    # 绿
    LinearSegmentedColormap.from_list(words[0], ['#f7fbff', '#4292c6'])     # 蓝
]

# ---------- 关键修改2：调整imshow的extent参数，确保每行高度完整 ----------
for i in range(len(words)):
    row_data = docs_matrix[i:i+1, :]
    im = ax2.imshow(row_data, cmap=row_cmaps[i], aspect='auto',
                   extent=[-0.5, len(years)-0.5, i+0.5, i-0.5],  # 修正行高范围（y轴上下限反转）
                   alpha=0.8)

# 设置坐标轴标签（注意顺序已反转）
ax2.set_title('Document Count Heatmap', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Term', fontsize=12)
ax2.set_xticks(np.arange(len(years)), labels=years)
ax2.set_yticks(np.arange(len(words)), labels=words)
# 调整Y轴范围确保顶部不遮挡
ax2.set_ylim(len(words)-0.5, -0.5)

# 添加数值标签
for i in range(len(words)):
    for j in range(len(years)):
        ax2.text(j, i, docs_matrix[i, j], ha='center', va='center', 
                color='black',
                fontsize=12)

# 调整布局
plt.tight_layout()
plt.show()
```
